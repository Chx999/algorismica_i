## Trobar el m√†xim

```python
def maxim_seq(A:list[int]) -> int:
    """
    Calcula el m√†xim d'una llista.
    :param
    A: llista d'enters
    :return
    un enter de la llista que √©s el valor m√†xim
    """
    maxim:int = A[0]
    for elem in A:
        if elem > maxim:
            maxim = elem
    return maxim
```

```python
maxim_seq([3,9,2,10,5])
```

```python
def maxim_div(A:list[int]) -> int:
    return maxim_div_rec(A,0,len(A))

def maxim_div_rec(A:list[int],inici,fi) -> int:
    """
    Calcula el m√†xim d'una llista.
    :param
    A: llista d'enters
    :return
    un enter de la llista que √©s el valor m√†xim
    """
    if fi-inici == 1:
        return (A[inici])
    else:
        mig = (inici + fi) // 2
        max_esq = maxim_div_rec(A,inici,mig)
        max_dret = maxim_div_rec(A,mig,fi)
        if max_esq > max_dret:
            return max_esq
        else:
            return max_dret
```

```python
c = maxim_div([3,9,5]) 
print("maxim",c)    
```

## Mergesort i la seva funci√≥ auxiliar, merge

```python
def merge(x:list[any],y:list[any])->list[any]:
    """
    Fusiona dues llistes pr√®viament ordenades en una nova llista ordenada
    :param
    x: una llista ordenada
    y: una llista ordenada
    :return
    una nova llista amb tots els elements d'x i y, ordenats
    """
    # print("MERGE: m'han cridat amb",x,y)  # xivato
    if len(x) < 1:
        return y
    if len(y) < 1:
        return x
    if x[0] <= y[0]:
        return [x[0]] + merge(x[1:],y)
    else:
        return [y[0]] + merge(x,y[1:])
```

```python
merge([],[4,5])
```

```python
merge([6,10],[])
```

```python
merge([3,7,10],[5,8,15])
```

```python
def mergesort(l:list[any])->list[any]:
    """
    Ordena una llista
    :param
    l1: una llista a ordenar
    :return
    una nova llista ordenada
    """
    # print("MERGESORT: m'han cridat amb",l)  # xivato
    if len(l) < 2:
        return l
    else:
        meitat = len(l) // 2
        esquerra = mergesort(l[:meitat])
        dreta = mergesort(l[meitat:])
        return merge(esquerra, dreta)
```

```python
mergesort([1,5,8,3,4,2,1,0])
```

### merge no recursiu

```python
def merge(x:list[any],y:list[any])->list[any]:
    """
    Fusiona dues llistes pr√®viament ordenades en una nova llista ordenada
    :param
    x: una llista ordenada
    y: una llista ordenada
    :return
    una nova llista amb tots els elements d'x i y, ordenats
    """
    resultat:list[any] = []
    i:int = 0
    j:int = 0
    while(i < len(x) and j < len(y)):
        if (x[i] <= y[j]):
            resultat.append(x[i])
            i = i + 1
        else:
            resultat.append(y[j])
            j = j + 1
    resultat += x[i:]
    resultat += y[j:]
    return resultat
```

```python
merge([3,7,10],[5,8,15])
```

```python
def quick_sort(A:list[any]):
    quick_sort_r(A, 0, len(A) - 1)    
                                      
def quick_sort_r(A:list[any], primer:int, darrer:int):  
    if darrer > primer:
        pivot:int = partition(A, primer, darrer) 
        quick_sort_r(A, primer, pivot - 1) 
        quick_sort_r(A, pivot + 1, darrer)  
```
