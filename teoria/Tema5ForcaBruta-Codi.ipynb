## üéí Algorisme de la motxilla

```python
from itertools import combinations
def motxilla_fb(nombre:int, capacitat:int, pes_cost:list[any]) -> (int,list[int]):
    """
    :param nombre: nombre d'items possibles
    :param capacitat: capacitat de la motxilla
    :param pes_cost: llista de tuples del tipus [(pes, cost), (pes, cost), ...]
    """
    millor_cost:int = None
    millor_combinacio:list[any] = []
    # genera totes les combinacions possibles d'1 o m√©s elements
    for num_elements in range(nombre):
        for comb in combinations(pes_cost, num_elements + 1):
            pes:int = sum([pc[0] for pc in comb])
            cost:int = sum([pc[1] for pc in comb])
            if (millor_cost is None or millor_cost < cost) and pes <= capacitat:
                millor_cost = cost
                millor_combinacio = [0] * nombre
                for pc in comb:
                    millor_combinacio[pes_cost.index(pc)] = 1
    return millor_cost, millor_combinacio
```

```python
motxilla_fb(4, 64, [(8,50),(16,100),(32,150),(40,200)])
```

```python
def johnson_trotter(n:int)->list[list[int]]:
    """
    Genera totes les N! permutacions del conjunt {1, 2, ..., n}.
    """
    # 1. Inicialitzaci√≥
    p:list[int] = list(range(1, n + 1))  
    direccions:list[bool] = [False] * n  # Direcci√≥: False = esquerra (<-); True = dreta (->)
    permutacions:list[list[int]] = [list(p)] 
    hi_ha_moviment:bool = True

    # 2. Bucle de Generaci√≥
    while hi_ha_moviment:

        # A. Trobar l'element M√≤bil m√©s gran
        element_mobil:int = -1
        index_mobil:int = -1
        trobat_mobil:bool = False
        
        for i in range(n):
            k = p[i]         
            adj_index:int = i - 1 if direccions[i] == False else i + 1
            if 0 <= adj_index < n and k > p[adj_index] and k > element_mobil:
                element_mobil = k
                index_mobil = i
                trobat_mobil = True
        hi_ha_moviment = trobat_mobil

        # B. Intercanvi: Moure l'element m√≤bil
        target_index:int = index_mobil - 1 if direccions[index_mobil] == False else index_mobil + 1          
        p[index_mobil], p[target_index] = p[target_index], p[index_mobil]
        direccions[index_mobil], direccions[target_index] = direccions[target_index], direccions[index_mobil]
        
        # C. Canvi de Direcci√≥: Revertir les direccions dels elements m√©s grans       
        for i in range(n):
            if p[i] > element_mobil:
                direccions[i] = not direccions[i]
                
        # Afegir la nova permutaci√≥
        permutacions.append(list(p))

    return permutacions
```

```python
johnson_trotter(4)
```

## ‚ôüÔ∏è Algoritme de les N Reines

```python
from itertools import permutations

def hi_ha_atac_diagonal(reines_per_columna:list[list[int]])->bool:
    """
    Comprova si dues reines s'ataquen diagonalment.
    'reines_per_columna' √©s una tupla (r0, r1, r2, ...), on ri √©s la fila de la reina a la columna i.
    """
    N:int = len(reines_per_columna)
    
    # Comprova totes les parelles de reines (i, j)
    for i in range(N):
        for j in range(i + 1, N):
            # i i j s√≥n les columnes
            # reines_per_columna[i] i reines_per_columna[j] s√≥n les files
            
            # Condici√≥ d'atac diagonal: |fila_i - fila_j| == |columna_i - columna_j|
            if abs(reines_per_columna[i] - reines_per_columna[j]) == abs(i - j):
                return True # Atac trobat
    return False

def n_reines_forca_bruta(N:int)->list[list[int]]:
    # Genera totes les permutacions de files (0 a N-1). 
    # Aix√≤ garanteix una reina per fila i una per columna.
    files:list[int] = range(N)
    totes_disposicions:list[list[int]] = permutations(files)
    solucions:int = 0

    # Iterem sobre totes les disposicions (FOR√áA BRUTA)
    for disposicio_files in totes_disposicions:
        # Si NO hi ha atac diagonal, √©s una soluci√≥ v√†lida
        if not hi_ha_atac_diagonal(disposicio_files):
            solucions += 1
            print(f"Soluci√≥ trobada: {disposicio_files}") # Descomentar per veure les solucions

    return solucions

```

```python
# Exemple d'√∫s (N=4 t√© 2 solucions)
N:int = 4
num_solucions:int = n_reines_forca_bruta(N)
print(f"N Reines (For√ßa Bruta de Permutaci√≥) per a N={N}: {num_solucions} solucions.")
```
