## üß† Codi per exemplificar el cost dels algorismes de text

```python
def is_unique(paraula:str) -> bool:
  caracters_vistos:list[str] = []
  for car in paraula:
    if car in caracters_vistos:
      return False
    caracters_vistos.append(car)
  return True
```

## üß† Algorisme de for√ßa bruta per cercar un text

```python
def CercaTextForcaBruta(t:str,p:str) -> int:  # conegut com BFStringMatching
    m:int = len(p)
    n:int = len(t)
    for i in range(n-m+1):                    # i √©s la posici√≥ inicial del patr√≥
      j:int = 0
      while j < m and p[j] == t[i+j]: 
        j += 1                                # j s√≥n els car√†cters que coincideixen
      if j == m: 
        return i      
    return -1
```

```python
def CercaTextForcaBruta(patro:str, text:str) -> int:  
    m:int = len(patro)                        # 1 pas
    n:int = len(text)                         # 1 pas
    if m > n:                                 # 1 pas 
        return -1                             # 1 pas
    for i in range(n-m+1):                    # el de dintre per n-m vegades 
      j:int = 0                               # 1 pas
      while j < m and patro[j] == text[i+j]:  # en el cas pitjor, el de dintre per m vegades
        j += 1                                # 2 passos
      if j == m:                              # 1 pas
        return i                              # 1 pas  (nom√©s s'executar√† un cop)
    return -1                                 # 1 pas (s'executar√† aquest o l'anterior)

# cost: 4 + (n-m)*(1 + (m-1)*2 + 1 ) + 1 => si nom√©s tenim en compte els components de m√©s pes  O(n*m)
```

## üß† Algorisme de Horspool

```python
def BoyerMooreHorspool(patro:str, text:str)->int:
    m:int = len(patro)
    n:int = len(text)
    if m > n: 
        return -1
    taula_desplacaments = []                             # creem la taula buida
    for k in range(256):                                 # car√†cters ASCII 
        taula_desplacaments.append(m)                    # per defecte el salt ser√† la llargada del patr√≥
    for k in range(m - 1):                            
        taula_desplacaments[ord(patro[k])] = m - k - 1  #  per als car√†cters del patr√≥ el salt √©s la distancia al final
                                                          # si apareixen m√©s d'un cop es guarda el darrer nom√©s
    taula_desplacaments = tuple(taula_desplacaments)      # convertim en tupla perqu√® no ho tocarem 
    k = m - 1                                             # k √©s la posici√≥ del text que comparem amb la lletra final
                                                          # inicialment equival a la darrera lletra del patr√≥
    while k < n:
        j = m - 1                                         # √≠ndex de la lletra del patr√≥ que comparem 
        i = k                                             # √≠ndex de la lletra del text que comparem
        while j >= 0 and text[i] == patro[j]:           
            j -= 1 
            i -= 1
                                                           # avancem(!) mentre hi hagi text i les lletres coincideixen
        if j == -1:                                        # si el patr√≥ s'ha acabat (totes les lletres han coincidit)  
            return i + 1                                   # retornem la posici√≥ del text
        k += taula_desplacaments[ord(text[k])]             # desplacem sense tenir en compte els aven√ßos fets
    return -1                                              # si hem arribat fins aqu√≠ no hem trobat el patr√≥
```

## üß† Algorisme de Levenshtein

```python
def distancia_levenshtein(patro, text):
    if len(text) == 0: 
        return len(patro)
    llargada_patro = len(patro) + 1
    llargada_text = len(text) + 1
    matriu_distancies = [[0] * llargada_text for x in range(llargada_patro)]
    for i in range(llargada_patro): 
        matriu_distancies[i][0] = i
    for j in range(llargada_text): 
        matriu_distancies[0][j] = j
    for i in range(1, llargada_patro):
        for j in range(1, llargada_text):
            deletion = matriu_distancies[i-1][j] + 1
            insercio = matriu_distancies[i][j-1] + 1
            substitucio = matriu_distancies[i-1][j-1]
            if patro[i-1] != text[j-1]: 
                substitucio += 1
            matriu_distancies[i][j] = min(insercio,deletion,substitucio)
    return matriu_distancies[llargada_patro-1][llargada_text-1]
```

## üß† Exemple d'algorisme amb finestra lliscant. Subcadena m√©s llarga

```python
def subcadena_mes_llarga(cadena:str)->str:
    """
    Aquesta funci√≥ identifica la subcadena m√©s llarga
    sense cap car√†cter repetit.
    Parameters
    ----------
    cadena: string
    Cadena donada
    Returns
    -------
    subcadena: string
    Subcadena m√©s llarga sense car√†cters repetits
    """
    n:int = len(cadena)
    # Fem un diccionari amb els car√†cters.
    diccCaracters:dict[str:str] = {caracter: False for caracter in cadena}
    # La soluci√≥ es basa en una finestra que es redefineix cada cop que troba un car√†cter repetit.
    iniciFinestra:int = 0
    # Inici i final de la finestra. El final √©s el darrer car√†cter vist; 
    # el principi √©s el primer car√†cter no repetit des del final de la finestra.
    iniciSubcadena:int = 0
    finalSubcadena:int = 0
    # Inici i final de la subcadena m√©s llarga trobada fins al moment
    for finalFinestra in range(0, n):
        if diccCaracters[cadena[finalFinestra]]: # Si el caracter ja hi era.
            while (cadena[iniciFinestra]!= cadena[finalFinestra]):
                diccCaracters[cadena[iniciFinestra]] = False
                iniciFinestra += 1
                # Desplacem la finestra a partir de la primera aparici√≥ 
                # del car√†cter, sense incloure'l.
            iniciFinestra += 1
        else:
            diccCaracters[cadena[finalFinestra]] = True
            # Anotem el car√†cter com a existent.
            finalFinestra += 1
            # Com que no hi ha repeticions augmentem la finestra.
            if finalSubcadena - iniciSubcadena < finalFinestra - iniciFinestra:
                # Revisem que la nova subcadena no sigui m√©s llarga que la que ten√≠em guardada.
                iniciSubcadena = iniciFinestra
                finalSubcadena = finalFinestra
    return cadena[iniciSubcadena:finalSubcadena]
```
